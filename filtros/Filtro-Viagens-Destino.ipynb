### Filtro de Viagens a Pé
Esse filtro considera apenas viagens com o transporte principal sendo "a pé", separando em zonas de acordo com seu destino

```python
import pandas as pd
```

#### Importando dados originais

```python
od = pd.read_csv('../../dados-od/OD_2017_v1.csv')
od.head()
```

#### Deixando apenas as viagens com o modo de transporte principal sendo "a Pé"

```python
od_walk = od.drop(od[od['MODOPRIN'] != 16.0].index, inplace=False)
od_walk.head()
```

#### Salvando os dataframes

```python
FILTER_NAME = "subprefeituras-destino"

for area, walks in filtered_walks.items():
    area = area.replace(' ', '_').replace('/', '_')
    walks.to_csv(f'../../dados/{FILTER_NAME}/{area}.csv')
```
